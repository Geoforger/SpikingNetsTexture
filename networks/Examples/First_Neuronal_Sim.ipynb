This file contains the code used to simulate a single Izhikevich neuron within NEST, utilising the PyNEST API

```python
# Import necessary libraries
import nest
import nest.voltage_trace   # Voltage trace output library
```

```python
# Set parameters for nest simulator environment
nest.set_verbosity("M_WARNING") # Set simulator to only output whenever a warning is encountered
nest.ResetKernel()  # Reset kernal to prevent errors
```

```python
# Create neuronal models
neuron = nest.Create("izhikevich")  # Create single izhikevich neuron
voltmeter = nest.Create("voltmeter")    # Voltmeter to show output
```

```python
# Set parameters for neurons
nest.SetStatus(neuron, "I_e", 200.0) # I_e gives the constant input current to the neuron
nest.SetStatus(voltmeter, [{"withgid": True}])
```

```python
# Connect neuron to voltmeter
nest.Connect(voltmeter, neuron)
```

```python
# Run simulation
nest.Simulate(10.0)
```

```python
# Get trace from voltmeter for output
nest.voltage_trace.from_device(voltmeter)
```
